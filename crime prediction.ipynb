```python
import sklearn
from sklearn.utils import shuffle
from sklearn.neighbors import KNeighborsClassifier
from sklearn.tree import DecisionTreeClassifier
from sklearn import svm
from sklearn.naive_bayes import GaussianNB
from sklearn.linear_model import LogisticRegression
from sklearn.neural_network import MLPClassifier
import pandas as pd
import numpy as np
from sklearn import linear_model, preprocessing
from sklearn.model_selection import train_test_split
import pickle
import warnings
warnings.filterwarnings('ignore')
```

```python
df = pd.read_csv (r'C:\Users\Lb\Downloads\DATA-BOOK-1-CI-2014-2015 (1).xlsx - INDEX.csv')
for i in df.columns:
    df[f"{i}"] = df[f"{i}"].str.replace(' ', '')
df.loc[(df["S_SEX"] == "UNKNOWN")|(df["S_SEX"] == "UNIDENTIFIED"), "S_SEX"] = "UNDETERMINED"
df.loc[df["V_SEX"] == "NOTSPECIFIED", "V_SEX"] = "UNDETERMINED"
df["S_AGE"]=df["S_AGE"].replace({"UNKNOWN": "0", "UNDETERMINED": "0", "NOTSPECIFIED": "0", "UNIDENTIFIED":"0","29LA":"29"})
df["V_AGE"]=df["V_AGE"].replace({"UNKNOWN": "0", "UNDETERMINED": "0", "NOTSPECIFIED": "0","UNIDENTIFIED":"0","3'6":"4"})
df.loc[(df["S_AGE"] == "LEGALAGE")|(df["S_AGE"] == "L")|(df["S_AGE"] == "LA"), "S_AGE"] = "-1"
df.loc[(df["V_AGE"] == "LEGALAGE")|(df["V_AGE"] == "L")|(df["V_AGE"] == "LA"), "V_AGE"] = "-1"

df.drop(df[df["S_AGE"].isna()].index, inplace = True)
df.drop(df[df["V_AGE"].isna()].index, inplace = True)
df.loc[(df['LAW_VIOLATED'].str.contains('9165')), 'LAW_VIOLATED'] = 'COMPREHENSIVE DANGEROUS DRUGS ACT OF 2002'
df.loc[(df['LAW_VIOLATED'].str.contains('9262')), 'LAW_VIOLATED'] = 'ANTI-VIOLENCE AGAINST WOMEN AND THEIR CHILDREN ACT OF 2004'
df.loc[(df['LAW_VIOLATED'].str.contains('9287')), 'LAW_VIOLATED'] = 'ACT TO INCREASE PENALTIES AGAINST ILLEGAL NUMBERS GAMES'
df["PLACE"] = df["PLACE"].str.replace('SANTAROSA','SANTAROSA',regex=True)
df.loc[(df['PLACE'].str.contains('SANTAROSA'))|(df['PLACE'].str.contains('SANATAROSA'))|
       (df['PLACE'].str.contains('STAROSA')), 'PLACE'] = 'SANTAROSA'

df.loc[(df['PLACE'].str.contains('CALAMB'))|(df['PLACE'].str.contains('CALANBA'))|
       (df['PLACE'].str.contains('CALABA')) , 'PLACE'] = 'CALAMBA'
df.loc[df['PLACE'].str.contains('BANLIC'), 'PLACE'] = 'CALAMBA'
df.loc[(df['PLACE'].str.contains('ALAMINOS'))|(df['PLACE'].str.contains('BRGY3POBALAM')), 'PLACE'] = 'ALAMINOS'
df.loc[(df['PLACE'].str.contains('CBUYAO'))|(df['PLACE'].str.contains('CABUYAO')), 'PLACE'] = 'CABUYAO'
df.loc[df['PLACE'].str.contains('MAJAYJAY'), 'PLACE'] = 'MAJAYJAY'
df.loc[(df['PLACE'].str.contains('LUMBAN'))|(df['PLACE'].str.contains('LUBAN')) , 'PLACE'] = 'LUMBAN'
df.loc[df['PLACE'].str.contains('PILA'), 'PLACE'] = 'PILA'
df.loc[(df['PLACE'].str.contains('SANPABLO'))|(df['PLACE'].str.contains('SANPBLO'))|
       (df['PLACE'].str.contains('SNAPABLO')) , 'PLACE'] = 'SANPABLO'
df.loc[df['PLACE'].str.contains('CAVINTI'), 'PLACE'] = 'CAVINTI'
df.loc[df['PLACE'].str.contains('CAVINITI'), 'PLACE'] = 'CAVINTI'
df.loc[df['PLACE'].str.contains('LILIW'), 'PLACE'] = 'LILIW'
df.loc[df['PLACE'].str.contains('PAGSANJAN'), 'PLACE'] = 'PAGSANJAN'
df.loc[df['PLACE'].str.contains('BRGYSABANGPAG'), 'PLACE'] = 'PAGSANJAN'
df.loc[df['PLACE'].str.contains('BINAN'), 'PLACE'] = 'BINAN'
df.loc[df['PLACE'].str.contains('BIÑAN'), 'PLACE'] = 'BINAN'
df.loc[df['PLACE'].str.contains('VICTORIA'), 'PLACE'] = 'VICTORIA'
df.loc[df['PLACE'].str.contains('MABITAC'), 'PLACE'] = 'MABITAC'
df.loc[df['PLACE'].str.contains('RIZAL'), 'PLACE'] = 'RIZAL'
df.loc[df['PLACE'].str.contains('RUZ'), 'PLACE'] = 'SANTACRUZ'
df.loc[df['PLACE'].str.contains('CALAUAN'), 'PLACE'] = 'CALAUAN'
df.loc[df['PLACE'].str.contains('NAGCARLAN'), 'PLACE'] = 'NAGCARLAN'
df.loc[df['PLACE'].str.contains('NAGARCLAN'), 'PLACE'] = 'NAGCARLAN'
df.loc[df['PLACE'].str.contains('KALAYAAN'), 'PLACE'] = 'KALAYAAN'
df.loc[df['PLACE'].str.contains('ANPEDRO'), 'PLACE'] = 'SANPEDRO'
df.loc[df['PLACE'].str.contains('SANPEDOR'), 'PLACE'] = 'SANPEDRO'
df.loc[df['PLACE'].str.contains('SINILOA'), 'PLACE'] = 'SINILOAN'
df.loc[df['PLACE'].str.contains('LOSBANO'), 'PLACE'] = 'LOSBANOS'
df.loc[df['PLACE'].str.contains('MAGDALENA'), 'PLACE'] = 'MAGDALENA'
df.loc[df['PLACE'].str.contains('PAKIL'), 'PLACE'] = 'PAKIL'
df.loc[df['PLACE'].str.contains('PANGIL'), 'PLACE'] = 'PANGIL'
df.loc[df['PLACE'].str.contains('BAY'), 'PLACE'] = 'BAY'
df.loc[df['PLACE'].str.contains('FAMY'), 'PLACE'] = 'FAMY'
df.loc[df['PLACE'].str.contains('BRGYBULIHANFIREARMMY'), 'PLACE'] = 'FAMY'
df.loc[df['PLACE'].str.contains('SANTAMARIA'), 'PLACE'] = 'SANTAMARIA'
df.loc[df['PLACE'].str.contains('LUISIANA'), 'PLACE'] = 'LUISIANA'
df.loc[df['PLACE'].str.contains('LUISIAN'), 'PLACE'] = 'LUISIANA'
df.loc[df['PLACE'].str.contains('PAETE'), 'PLACE'] = 'PAETE'
df.loc[df['PLACE'].str.contains('STAMARIA'), 'PLACE'] = 'SANTAMARIA'
df.loc[df['PLACE'].str.contains('BRGYSANTISIMOROSARIO'), 'PLACE'] = 'SANPABLO'
df.loc[df['PLACE'].str.contains('BRGYSANFRANCISCOSNAPBLOCITY'), 'PLACE'] = 'SANPABLO'
df.loc[df['PLACE'].str.contains('BRGYPOBLACIONMAGALENA'), 'PLACE'] = 'MAGDALENA'
df.loc[df['PLACE'].str.contains('LOBANOS'), 'PLACE'] = 'LOSBANOS'
df.loc[df['PLACE'].str.contains('BINAM'), 'PLACE'] = 'BINAN'
df.loc[df['PLACE'].str.contains('BRGYLUCONGMBITAC'), 'PLACE'] = 'MABITAC'
df.loc[df['PLACE'].str.contains('NAGCARAN'), 'PLACE'] = 'NAGCARLAN'
df.loc[df['PLACE'].str.contains('LOSBAÑOS'), 'PLACE'] = 'LOSBANOS'
df.loc[df['PLACE'].str.contains('#VALUE!'), 'PLACE'] = 'UNDETERMINED'
df = df[(df["LAW_VIOLATED"] == "THEFT") | (df["LAW_VIOLATED"] == "PHYSICAL INJURY") | 
(df["LAW_VIOLATED"] == "ROBBERY") | (df["LAW_VIOLATED"] == "MURDER") | (df["LAW_VIOLATED"] == "RAPE") | 
(df["LAW_VIOLATED"] == "HOMICIDE")|(df["LAW_VIOLATED"] == 'CARNAPPING')
|(df["LAW_VIOLATED"] == 'ACT TO INCREASE PENALTIES AGAINST ILLEGAL NUMBERS GAMES')
|(df["LAW_VIOLATED"] == 'COMPREHENSIVE DANGEROUS DRUGS ACT OF 2002')|
(df["LAW_VIOLATED"] == 'ANTI-VIOLENCE AGAINST WOMEN AND THEIR CHILDREN ACT OF 2004')]
df["V_AGE"] = pd.to_numeric(df["V_AGE"])
df["S_AGE"] = pd.to_numeric(df["S_AGE"])
```

```python
crimes = {}
for i in df["LAW_VIOLATED"].unique():
    crimes[i] = len(df[df["LAW_VIOLATED"] == i])
    
print({k: v for k, v in sorted(crimes.items(), key=lambda item: item[1])})
```

```python
le = preprocessing.LabelEncoder()
# df["time"] = le.fit_transform(df["TIME"])
# df["v_nationality"] = le.fit_transform(df["V_NATIONALITY"])
df["v_age"] = (df["V_AGE"])
df["s_age"] = (df["S_AGE"])
df["v_sex"] = le.fit_transform(df["V_SEX"])

keys = le.classes_
values = le.transform(le.classes_)
dictionary = dict(zip(keys, values))
print("v_sex",dictionary)
print()

df["s_sex"] = le.fit_transform(df["S_SEX"])

keys = le.classes_
values = le.transform(le.classes_)
dictionary = dict(zip(keys, values))
print("s_sex",dictionary)
print()

df["s_use_of"] = le.fit_transform(df["S_USE_OF"])

keys = le.classes_
values = le.transform(le.classes_)
dictionary = dict(zip(keys, values))
print("s_use_of",dictionary)
print()

df["place"] = le.fit_transform(df["PLACE"])

keys = le.classes_
values = le.transform(le.classes_)
dictionary = dict(zip(keys, values))
print("place",dictionary)
print()

df["law_violated"] = le.fit_transform(df["LAW_VIOLATED"])

keys = le.classes_
values = le.transform(le.classes_)
dictionary = dict(zip(keys, values))
print("law_violated",dictionary)
print()

# df["s_nationality"] = le.fit_transform(df["S_NATIONALITY"])
# df["s_status"] = le.fit_transform(df["S_STATUS"])
# df["actions_taken"] = le.fit_transform(df["ACTIONS_TAKEN"].astype(str))
# df["remarks"] = le.fit_transform(df["REMARKS"].astype(str))
df = df.drop(["S_AGE","V_AGE","BLOTTER_ENTRY","DATE","TIME","PLACE","LAW_VIOLATED","V_SEX","V_NATIONALITY","S_USE_OF","S_SEX","S_NATIONALITY","S_STATUS","ACTIONS_TAKEN","REMARKS"], axis="columns")
df
```

```python
keys = le.classes_
values = le.transform(le.classes_)
dictionary = dict(zip(keys, values))
print(dictionary)
```

```python
label = "law_violated"
x = np.array(df.drop([label], axis="columns"))
y = np.array(df[label])
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size = 0.25)


mlpc= MLPClassifier(solver='lbfgs', alpha=1e-5, hidden_layer_sizes=(15,),  warm_start=True, random_state=789)
lr = LogisticRegression(random_state=0)
gnb = GaussianNB()
knn = KNeighborsClassifier(n_neighbors=23)
dt = DecisionTreeClassifier(random_state=244)
_svm = svm.SVC(random_state=0, probability=True)

mlpc.fit(x_train, y_train)
lr.fit(x_train, y_train)
gnb.fit(x_train, y_train)
knn.fit(x_train, y_train)
dt.fit(x_train, y_train)
_svm.fit(x_train, y_train)

mlpc_acc = mlpc.score(x_test, y_test)
lr_acc = lr.score(x_test, y_test)
gnb_acc = gnb.score(x_test, y_test)
knn_acc = knn.score(x_test, y_test)
dt_acc = dt.score(x_test, y_test)
_svm_acc = _svm.score(x_test, y_test)
    

print("mlpc_acc",mlpc_acc)
print("lr_acc",lr_acc)
print("gnb_acc",gnb_acc)
print("knn_acc",knn_acc)
print("dt_acc",dt_acc)
print("_svm_acc",_svm_acc)
```
